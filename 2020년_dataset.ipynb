```python
import pandas as pd

df = pd.read_csv('Data\시도별 온실가스 중립도 (2015_2020년).csv', low_memory=False)

df1 = df[df['년도']==2020]
# df1['온실가스흡수량'] = df1['온실가스흡수량'].astype('float64')
df1['온실가스흡수량'] = df1['온실가스흡수량'].astype('int64')
df1
```

```python
import pandas as pd

df1['시도명'] = df1['시도명'].replace({
    '강원도': '강원',
    '경기도': '경기',
    '경상남도': '경남',
    '경상북도': '경북',
    '광주광역시': '광주',
    '대구광역시': '대구',
    '대전광역시': '대전',
    '부산광역시': '부산',
    '서울특별시': '서울',
    '세종특별자치시': '세종',
    '울산광역시': '울산',
    '인천광역시': '인천',
    '전라남도': '전남',
    '전라북도': '전북',
    '제주특별자치도': '제주',
    '충청남도': '충남',
    '충청북도': '충북'
})
```

```python
df1
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt


font_path = "c:\Windows\Fonts\malgun.ttf"  
font_prop = mpl.font_manager.FontProperties(fname=font_path) 
mpl.rcParams['font.family'] = font_prop.get_name()

df1_sorted1 = df1.sort_values(by='온실가스중립도', ascending=False)

plt.figure(figsize=(17,5))
plt.bar(df1_sorted1['시도명'],df1_sorted1['온실가스중립도'])
plt.title('시도별 온실가스 중립도')
plt.xlabel('시도명')
plt.ylabel('온실가스 중립도 (%)')
plt.xticks(rotation=45) 
plt.show()
```
