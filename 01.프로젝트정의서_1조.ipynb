# 시도별 온실가스 중립율(흡수량/발생량) 분석 및 개선방안 고찰

탄소중립이 강조되는 현재에, 우리나라의 온실가스 배출 및 흡수량 파악을 통해서, 환경보존의 지속가능성을 평가하고,
현황 분석을 통한 지역별 온실가스 흡수량을 증가시키는 개선방을 고찰을 위한 프로젝트이다.

국가기관에서 제공하는 데이터셋을 통하여
시도별 온실가스 배출 및 흡수량을 비교하여 각 지역별 온실가스 중립율을 파악.
지역별 온길가스 중립율에 대한 현황을 진단.
국가 목표수치인 온실가스 중립율 40% 달성을 위한 온실가스 흡수량 증대를 위한 개선방안 고찰

- **기간:** 2024-08-12 ~ 2024-08-22 (총 7일)
- **정의:**
  - 이번 프로젝트는 대한민국의 탄소중립 현황을 분석하고, 2050 탄소중립 정책 발표 이후 온실가스 배출 및 흡수량의 연도별 변화를 검토
  - 탄소중립 목표인 40% 달성을 위한 정책의 실효성을 평가하는 것을 목표로 한다. 
- **목적:**
  - 국내 온실가스 배출량과 흡수량을 조사하여 지역별 온실가스 중립율을 파악
  - 국가의 목표수치인 탄소중립 40%의 달성을 위한 정책의 실효성 검토 및 정책 제언
- **범위:**
  - 17년부터 21년까지의 최근 5개년을 대상으로 실시
  - 지역별 온실가스 배출 및 흡수량에 대한 데이터 분석
  - 21년 탄소중립2050 발표 후 현재까지의 국내 정책 검토

- **결론 및 향후 과제:**
  - 온실가스 배출량 및 흡수량에 대한 추이 확인
  - 대다수의 국내 정책이 온실가스 배출량의 감소를 위한 노력 및 투자하며 흡수량에 대한 발전노력은 상대적 미비
  - '탄소배출거래제도'를 통하여 온실가스 흡수량에 대한 효율적인 소비(?)
  - 지역별 특수성으로 인한 일부지역의 정책 유리
