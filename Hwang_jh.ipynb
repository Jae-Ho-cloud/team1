```python
import pandas as pd

df = pd.read_csv('Data\시도별 온실가스 중립도 (2015_2020년).csv', low_memory=False)

df1 = df[df['년도']==2020]
df1['온실가스흡수량'] = df1['온실가스흡수량'].astype('int64')
df1
```
